```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# X range
x_vals = np.linspace(start=-10., stop=10., num=100)
```

```python
import tensorflow as tf

# Relu Activation Function : (0,Max)
print(tf.nn.relu(5).numpy())
print(tf.nn.relu(-1).numpy())
print(tf.nn.relu(3.5).numpy())
print(tf.nn.relu(0).numpy())
print(tf.nn.relu(0.1).numpy())
```

```python
# Sigmoid : 1/ (1 + e-x)

print(tf.nn.sigmoid(1.0).numpy())
print(tf.nn.sigmoid(-0.5).numpy())
print(tf.nn.sigmoid(3.4).numpy())
print(tf.nn.sigmoid(-2.1).numpy())
print(tf.nn.sigmoid(0.0).numpy())
```

```python
#tanh : (ez - e-z) / (ez + e-z)

print(tf.nn.tanh(1.0).numpy())
print(tf.nn.tanh(-0.5).numpy())
print(tf.nn.tanh(3.4).numpy())
print(tf.nn.tanh(-2.1).numpy())
print(tf.nn.tanh(0.0).numpy())
print(tf.nn.tanh(-6.5).numpy())
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt

# Initialize the X range values for plotting
x_vals = tf.linspace(-10.,10.,100)

# Relu Activation Function
y_relu = tf.nn.relu(x_vals)

plt.plot(x_vals, y_relu, 'b:', label='ReLU', linewidth=2)
```

```python
# Relu6
# Relu6 Activation Function
y_relu6 = tf.nn.relu6(x_vals)

plt.plot(x_vals, y_relu, 'b:', label='ReLU', linewidth=2)

print("Relu 6 =",tf.nn.relu6([-3.,3.,10.]).numpy())
```

```python
# sigmoid
y_sigmoid = tf.nn.sigmoid(x_vals)
print(tf.nn.sigmoid([-1.,0.,1.]).numpy())
plt.plot(x_vals,y_sigmoid)
```

```python
# Hyper Tangent activation
y_hyperTanH = tf.nn.tanh(x_vals)
plt.plot(x_vals,y_hyperTanH)
```

```python
# Softsign activation
y_softsign = tf.nn.softsign(x_vals)
plt.plot(x_vals,y_softsign)
```

```python
# Softplus activation
y_softplus = tf.nn.softplus(x_vals)
plt.plot(x_vals,y_softsign)
```

```python
# # Exponential linear activation
y_elu = tf.nn.elu(x_vals)
plt.plot(x_vals,y_elu)
```
