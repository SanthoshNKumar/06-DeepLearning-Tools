### tile

```python
# This operation creates a new tensor by replicating input multiples times. 
# The output tensor's i'th dimension has input.dims(i) * multiples[i] elements, 
# and the values of input are replicated multiples[i] times along the 'i'th dimension. For example, 
# tiling [a b c d] by [2] produces [a b c d a b c d].

a = tf.constant([[1,2,3],[4,5,6]], tf.int32) 
b = tf.constant([1,2], tf.int32) 
print(tf.tile(a, b),"\n")

c = tf.constant([2,1], tf.int32) 
print(tf.tile(a, c),"\n")

d = tf.constant([2,2], tf.int32) 
print(tf.tile(a, d),"\n")
```

### Plotting 'L1' and 'L2' Loss

```python
import tensorflow as tf
import matplotlib.pyplot as plt

# Various Predicted X-values
x_vals = tf.linspace(-1.,1.,500)

target = tf.constant(0.) # Create our target of zero

# L2 Loss
# L = (pred - actual)^2
l2_y_out = tf.square(target - x_vals)

# L1 Loss
# L = abs(pred - actual)
l1_y_out = tf.abs(target - x_vals)

plt.plot(x_vals,l2_y_out,'b-',color='red',label="L2_Loss")
plt.plot(x_vals,l1_y_out,'r--',color='blue',label="L1_Loss")
plt.grid()
plt.legend(loc='lower right', prop={'size': 11})
plt.show()
```

```python
# Gradient Page

import tensorflow as tf
import numpy as np

x_vals = np.random.normal(1, 0.1, 100)
y_vals = np.repeat(10., 100)

# Create variable (one model parameter = W)
W = tf.Variable(tf.random.normal(shape=[1]))
B = tf.Variable(tf.random.normal(shape=[1]))

deltaW = tf.constant(0,tf.float32)
deltaB = tf.constant(0,tf.float32)

with tf.GradientTape() as tape:
    prediction = W * x_vals + B
    error = prediction - y_vals
    squaredError = tf.square(error)
    loss_ = tf.reduce_mean(input_tensor=squaredError)
    deltaW,deltaB = tape.gradient(loss_,[W,B])
    
changeDW = deltaW * 0.01
changeDB = deltaB * 0.01

print("Old W=",W.numpy())
print("deltaW =",np.array(deltaW))
print("changeDW",changeDW.numpy())
print("New W",W.assign_sub(changeDW).numpy())

print("\n")

print("Old B=",B.numpy())
print("deltaB =",np.array(deltaB))
print("changeDB",changeDB.numpy())
print("New B",B.assign_sub(changeDB).numpy())
```

```python
import tensorflow as tf
import numpy as np

x_vals = np.random.normal(1, 0.1, 100)
y_vals = np.repeat(10., 100)

# Create variable (one model parameter = W)
W = tf.Variable(tf.random.normal(shape=[1]))
B = tf.Variable(tf.random.normal(shape=[1]))

deltaW = tf.constant(0,tf.float32)
deltaB = tf.constant(0,tf.float32)
loss = tf.constant(0,tf.float32)

for i in range(400):
    
    with tf.GradientTape() as tape:
        prediction = W * x_vals + B
        error = prediction - y_vals
        squaredError = tf.square(error)
        loss = tf.reduce_mean(input_tensor=squaredError)
        deltaW,deltaB = tape.gradient(loss,[W,B])
        
    changeW = deltaW * 0.01
    changeB = deltaB * 0.01
    
    W.assign_sub(changeW)
    B.assign_sub(changeB)
    
    print("Iteration= {0},W={1} and B = {2} and Loss={3}".format(i,W.numpy(),B.numpy(),loss))
    
```

```python
# Stochastic Training:

import tensorflow as tf
import numpy as np

x_vals = np.random.normal(1, 0.1, 100)
y_vals = np.repeat(10., 100)

# Create variable (one model parameter = W)
W = tf.Variable(tf.random.normal(shape=[1]))
B = tf.Variable(tf.random.normal(shape=[1]))

deltaW = tf.constant(0,tf.float32)
deltaB = tf.constant(0,tf.float32)
loss = tf.constant(0,tf.float32)

for i in range(400):
    rand_index = np.random.choice(100)
    rand_x = [x_vals[rand_index]]
    rand_y = [y_vals[rand_index]]
    print("rand_index={0}, x_vals={1} and y_vals={2}".format(rand_index,rand_x,rand_y))
    
    with tf.GradientTape() as tape:
        prediction = W * rand_x + B
        error = (prediction - rand_y)
        squaredError = tf.square(error)
        loss = tf.reduce_mean(input_tensor=squaredError)
        deltaW,deltaB = tape.gradient(loss,[W,B])
        
    changeW = deltaW * 0.01
    changeB = deltaB * 0.01
    
    W.assign_sub(changeW)
    B.assign_sub(changeB)
    
    print("Iteration= {0},W={1} and B = {2} and Loss={3}".format(i,W.numpy(),B.numpy(),loss))
    
```

```python
# Batch Training

import tensorflow as tf
import numpy as np

# Declare batch size
batch_size = 20

x_vals = np.random.normal(1, 0.1, 100)
y_vals = np.repeat(10., 100)

# Create variable (one model parameter = W)
W = tf.Variable(tf.random.normal(shape=[1]))
B = tf.Variable(tf.random.normal(shape=[1]))

deltaW = tf.constant(0,tf.float32)
deltaB = tf.constant(0,tf.float32)
loss = tf.constant(0,tf.float32)

for i in range(400):
    rand_index = np.random.choice(100, size=batch_size)
    rand_x = np.transpose([x_vals[rand_index]])
    rand_y = np.transpose([y_vals[rand_index]])
    print("Trainig for the Batch values of :",rand_index)
    # print("rand_index={0}, x_vals={1} and y_vals={2}".format(rand_index,rand_x,rand_y))
    
    with tf.GradientTape() as tape:
        prediction = W * rand_x + B
        error = (prediction - rand_y)
        squaredError = tf.square(error)
        loss = tf.reduce_mean(input_tensor=squaredError)
        deltaW,deltaB = tape.gradient(loss,[W,B])
        
    changeW = deltaW * 0.01
    changeB = deltaB * 0.01
    
    W.assign_sub(changeW)
    B.assign_sub(changeB)
    
    print("Iteration= {0},W={1} and B = {2} and Loss={3}".format(i,W.numpy(),B.numpy(),loss))
    print("\n")
    
```
