```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# We create our data, placeholders and variables
x_val = np.random.normal(1, 0.1, 100)   #input values
y_val = np.repeat(10., 100)   #target values

#Variables
# Create variables for linear regression
w = tf.Variable(tf.random.normal(shape=[1,1]))
b = tf.Variable(tf.random.normal(shape=[1,1]))

loss = []
epochs = []

my_opt = tf.optimizers.SGD(0.001)

for i in range(100):
    with tf.GradientTape() as t:
        # Add linear function to the computational graph
        y_pred = tf.add(tf.multiply(x_val, w),b)
        current_loss = tf.reduce_sum(tf.square(y_pred - y_val))
        
    grads = t.gradient(current_loss, [w,b])
    my_opt.apply_gradients(zip(grads,[w,b]))
    loss.append(current_loss)
    epochs.append(i)
    
plt.plot(epochs,loss)
plt.grid()
plt.xlabel("Epoch")
plt.ylabel("Loss")
```
