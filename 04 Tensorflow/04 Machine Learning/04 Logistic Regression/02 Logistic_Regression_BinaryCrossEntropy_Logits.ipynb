```python
import numpy as np
import tensorflow as tf
from sklearn.datasets import load_iris

iris = load_iris()

binary_target = np.array([1. if x==0 else 0. for x in iris.target])
x1_data = [x[2] for x in iris.data]
x2_data = [x[3] for x in iris.data]

# Declare batch size
batch_size = 20

A = tf.Variable(tf.random.normal(shape=[1,1]))
b = tf.Variable(tf.random.normal(shape=[1,1]))

x1_data = tf.expand_dims(tf.constant(x1_data,tf.float32),1)
x2_data = tf.expand_dims(tf.constant(x2_data,tf.float32),1)

y = tf.expand_dims(tf.constant(binary_target,tf.float32),1)

# Create a model  = x1 - A*x2 + b
my_mult = tf.matmul(x2_data, A)
my_add = tf.add(my_mult, b)
my_output = tf.subtract(x1_data, my_add)
# my_output = tf.subtract(x_data[0], tf.add(tf.matmul(x_data[1], A), b))

# Add classification loss (cross entropy)
xentropy = tf.nn.sigmoid_cross_entropy_with_logits(labels=my_output, logits=y)

optimizer = tf.optimizers.SGD(0.05)
loss = []

for i in range(10):
    with tf.GradientTape() as t:
        my_mult = tf.matmul(x2_data, A)
        my_add = tf.add(my_mult, b)
        my_output = tf.subtract(x1_data, my_add)
        xentropy = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(labels=my_output, logits=y))
        
    grads = t.gradient(xentropy,[A,b])
    optimizer.apply_gradients(zip(grads,[A,b]))
    
    loss.append(xentropy.numpy())
    
    print("Step {0} : A= {1} and b = {2}".format(i,A.numpy(),b.numpy()))
```
